## AFAR: baseline and dummy benchmarks

```python
def compare(metric, y_label):
    data = []
    names = []
    scenarios = ['Baseline', 'Dummy', 'AFAR']
    colors = ['#512C96', '#519259', '#5800FF']
    for i in range(0, 4):
        data.append([baseline[metric][i], dummy[metric][i], approach_rank_1[metric][i]])
        names.append(baseline['table'][i])
        data[i] = [el * 100 for el in data[i]]
    
    fig = plt.figure(figsize=(9, 4.5))
    
    plt.subplot(141)
    plt.title('football')
    plt.bar(scenarios, data[0], width=0.95, color=colors)
    plt.tick_params(axis='x', bottom=False, labelbottom=False)
    plt.ylabel(y_label, fontsize=13)
    plt.ylim((50, 105))
    plt.yticks([50, 60, 70, 80, 90, 100])
    plt.grid(axis = 'y')
    
    plt.subplot(142)
    plt.title('kidney_disease')
    plt.bar(scenarios, data[1], width=0.95, color=colors)
    plt.tick_params(axis='x', bottom=False, labelbottom=False)
    plt.tick_params(axis='y', left=False, labelleft=False)
    plt.ylim((50, 105))
    plt.grid(axis = 'y')
    
    plt.subplot(143)
    plt.title('steel_plate_fault')
    plt.bar(scenarios, data[2], width=0.95,color=colors)
    plt.tick_params(axis='x', bottom=False, labelbottom=False)
    plt.tick_params(axis='y', left=False, labelleft=False)
    plt.ylim((50, 105))
    plt.grid(axis = 'y')
    
    plt.subplot(144)
    plt.title('titanic')
    plt.bar(scenarios, data[3], width=0.95, color=colors)
    plt.tick_params(axis='x', bottom=False, labelbottom=False)
    plt.tick_params(axis='y', left=False, labelleft=False)
    plt.ylim((50, 105))
    plt.grid(axis = 'y')
    
    baseline_patch = mpatches.Patch(color='#512C96', label='baseline')
    dummy_patch = mpatches.Patch(color='#519259', label='dummy')
    afa_patch = mpatches.Patch(color='#5800FF', label='AFAR Rank_1')
    fig.legend(bbox_to_anchor =(0.73, 1.03), handles=[baseline_patch, dummy_patch, afa_patch], ncol=3)
compare('accuracy', 'accuracy (%)')
```
